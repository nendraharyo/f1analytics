<H3><strong>Verstappen Melakukan Strategi Harvesting ERS Agar Dapat Menjauh dari Norris Saat Berada di Zona Deteksi DRS</strong><H3>

```python
#menampilkan grafik speed lap 61
fig,ax=plt.subplots(figsize=(14,7))
vertelem=imolagp_r.laps.pick_lap(61).pick_driver('VER').get_telemetry()
nortelem=imolagp_r.laps.pick_lap(61).pick_driver('NOR').get_telemetry()
ax.plot(vertelem['Distance'],vertelem['Speed'],c='#'+imolagp_r.get_driver('VER').TeamColor,label='VER')
ax.plot(nortelem['Distance'],nortelem['Speed'],c='orange',label='NOR')
ax.set_xticks(imolagp_r.get_circuit_info().corners['Distance'])
ax.set_xticklabels(imolagp_r.get_circuit_info().corners['Number'])
ax.legend()
ax.set_xlabel('Tikungan ke-')
ax.set_ylabel('Kecepatan (kpj)')
```

```python
#menampilkan grafik speed lap 62
fig,ax=plt.subplots(figsize=(14,7))
vertelem=imolagp_r.laps.pick_lap(62).pick_driver('VER').get_telemetry()
nortelem=imolagp_r.laps.pick_lap(62).pick_driver('NOR').get_telemetry()
ax.plot(vertelem['Distance'],vertelem['Speed'],c='#'+imolagp_r.get_driver('VER').TeamColor,label='VER')
ax.plot(nortelem['Distance'],nortelem['Speed'],c='orange',label='NOR')
ax.set_xticks(imolagp_r.get_circuit_info().corners['Distance'])
ax.set_xticklabels(imolagp_r.get_circuit_info().corners['Number'])
ax.legend()
ax.set_xlabel('Tikungan ke-')
ax.set_ylabel('Kecepatan (kpj)')
```

<li>
Dengan kesulitan Verstappen dalam menggunakan ban Hardnya, untuk mencegah Norris mendapatkan DRS Verstappen harus memaksimalkan harvesting ERS agar bisa menggunakan ERS untuk 'kabur' saat tiba di zona deteksi DRS
</li>
<li>
Harvesting ERS dilakukan dengan melakukan early braking sebelum tikungan kedua dan ke-14
</li>
