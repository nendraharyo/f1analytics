```python
import math
result={}
for drvs in miamigp_r.drivers:
    #fig,ax=plt.subplots()
    drv=miamigp_r.get_driver(drvs).Abbreviation
    tcolors='#'+miamigp_r.get_driver(drv).TeamColor
    test=miamigp_r.laps.pick_driver(drv).pick_accurate().pick_wo_box()[['LapTime','LapNumber']]
    test=test[test['LapNumber']>(57-17)]
    #test.plot(x='LapNumber',y='LapTime',ax=ax,label='RACE',color=tcolors)
    x=miamigp_s.laps.pick_driver(drv).pick_wo_box()[['LapTime','LapNumber']]
    x['LapNumberAdjusted']=x['LapNumber']+(57-19)
    df_sr=test.merge(x,left_on='LapNumber',right_on='LapNumberAdjusted',suffixes=['_race','_sprint'])[['LapTime_race','LapTime_sprint']]
    df_sr['gap']=df_sr['LapTime_race']-df_sr['LapTime_sprint']
    
    #x.plot(x='LapNumberAdjusted',y='LapTime',ax=ax,label='SPRINT',color=tcolors,linestyle=':')
    sec=df_sr['gap'].median().total_seconds()
    if math.isnan(sec)!=True:
        result[drv]=sec
```

```python
df=pd.DataFrame(result,index=[0]).T.sort_values(0,ascending=False)
sns.set_theme(rc={'figure.figsize':(15,9)})
fastf1.plotting.setup_mpl()
g=sns.barplot(data=df,x=0,y=df.index,hue=0,order=df.index[::-1],palette=sns.color_palette("coolwarm", as_cmap=True),legend=False)

for index, row in df.iterrows():
    if row[0]<0:
        g.text( row[0], index, round(row[0],3), color='white', ha="right")
    else:
        g.text( row[0], index, '+'+str(round(row[0],3)), color='white', ha="left")

g.set_ylabel('Pembalap')
g.set_xlabel('Rata-rata perbedaan waktu Race (17 Lap terakhir) dengan Sprint')
g.grid(c='grey')
g.set_title('Rata-Rata Pace Race (17 Lap terakhir) vs Sprint')
```
